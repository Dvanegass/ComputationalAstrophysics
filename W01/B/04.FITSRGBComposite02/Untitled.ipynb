```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.visualization import make_lupton_rgb
```

```python
hdulg = fits.open('data/NGC123g.fits')
hduli = fits.open('data/NGC123i.fits')
hdulr = fits.open('data/NGC123r.fits')
hdulu = fits.open('data/NGC123u.fits')
```

```python
i = hduli[0].data
g = hdulg[0].data
r = hdulr[0].data
u = hdulu[0].data
```

```python
u.shape
```

```python
plt.figure()
plt.imshow(r, cmap='gray',vmax=10)
plt.show()
```

```python
np.min(u)
```

```python
np.max(u)
```

```python
rgb_composite= make_lupton_rgb(r,g,u)

rgb_composite.shape
```

```python
plt.figure(figsize=(7,7))
plt.imshow(rgb_composite, origin='lower')
plt.show()
```

```python
rgb_composite= make_lupton_rgb(i,r,g, minimum=0, stretch=0.8, Q=18)
plt.figure(figsize=(7,7))
plt.imshow(rgb_composite, origin='lower')
plt.show()
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
%matplotlib inline
```

```python
hdul = fits.open('data/image0.fits')
hdul.info()
```

```python
hdul[0].header
```

```python
data = hdul[0].data
plt.figure()
plt.imshow(data)
plt.xlabel('x-pixel (RA)')
plt.ylabel('y-pixel (Dec)')
plt.colorbar()
plt.show()
```

```python
files = ['data/image0.fits', 'data/image1.fits', 'data/image2.fits', 'data/image3.fits']

data = np.zeros([4,200,200])

for i in range(4):
    hdul = fits.open(files[i])
    data[i] = hdul[0].data
    hdul.close()

data[0]
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].imshow(data[0])
ax[0,1].imshow(data[1])
ax[1,0].imshow(data[2])
ax[1,1].imshow(data[3])
plt.show()
```

```python

plt.figure()
plt.imshow(data[3])
plt.xlabel('x-pixel (RA)')
plt.ylabel('y-pixel (Dec)')
plt.colorbar()
plt.show()
```

```python
mean_data = np.mean(data, axis=0)

plt.figure()
plt.imshow(mean_data)
plt.xlabel('x-pixel (RA)')
plt.ylabel('y-pixel (Dec)')
plt.colorbar()
plt.show()
```

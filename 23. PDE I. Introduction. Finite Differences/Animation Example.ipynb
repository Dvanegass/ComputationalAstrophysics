```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
%matplotlib inline


#Here the function to be graphed is defined
def f(x,x0):
    sigma2 = 20.
    return np.exp(-(x-x0)**2/(2*sigma2))

# First the stage on which the animation is performed is created.
# In this case a figure and axes.
# In addition, the labels of the axes are written since they will not change
# throughout the animation
fig, ax=plt.subplots()
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(t,x)$')

# In this list the frames of the animation are stored
imgs = []
xrange = np.linspace(0,50,100)
#Here is the main loop.
# In each step a line is created with the function located at a value of x0
# and the title is written with the time of the frame
for i in xrange:
    line, = ax.plot(xrange,f(xrange, 30+i), color='crimson')#ax.plot(a[:,i])
    title = ax.text(2.,0.95,f'Time = {i:.2f}')
    # Here the image is added to the list
    imgs.append([line, title])

# The ArtistAnimation() function joins the images in the list to create the
# object with animation information
anim = animation.ArtistAnimation(fig, imgs, interval=50, blit=False)

# To display the animation in jupyter it must be converted to another format.
# For example which is converted to a javascript video
HTML(anim.to_jshtml())
```

```python
# transform to html5
HTML(anim.to_html5_video())
```

### Abstract

In this notebook we generate some synthetic image of the Sun and radomly distributed sunspots.

---

```python
import numpy as np
import matplotlib.pyplot as plt


def genImg():
    fuz = 0.5
    image_data = np.ones([28,28])
    
    
    N = np.random.randint(6)
    i=0
    for x in range(28):
        for y in range(28):
            if (x-14)**2 + (y-14)**2 >= 100:
                image_data[x,y] = 0.8
                
    for x in range(28):
        for y in range(28):
            if (x-14)**2 + (y-14)**2 >= 121:
                image_data[x,y] = 0.6

    for x in range(28):
        for y in range(28):
            if (x-14)**2 + (y-14)**2 >= 144:
                image_data[x,y] = 0
                
    while i<N:
        x,y = np.random.randint(28, size=2)
        if (x-14)**2 + (y-14)**2 < 100:
            image_data[x,y] = 0
            image_data[x-1,y] = np.random.rand()*fuz+0.5
            image_data[x-1,y-1] = np.random.rand()*fuz+0.5
            image_data[x-1,y+1] = np.random.rand()*fuz+0.5
            image_data[x,y-1] = np.random.rand()*fuz+0.5
            image_data[x,y+1] = np.random.rand()*fuz+0.5
            image_data[x+1,y] = np.random.rand()*fuz+0.5
            image_data[x+1,y-1] = np.random.rand()*fuz+0.5
            image_data[x+1,y+1] = np.random.rand()*fuz+0.5
            i +=1
    
    return image_data, N


data, target = genImg()
print(target)
plt.figure(figsize=(2,2))
plt.imshow(data, cmap='gray')
plt.show()
```

```python
# Save training data
samples = 5000
data = np.zeros([samples,28,28])
targets = np.zeros(samples, dtype=int)
for j in range(samples):
    data[j], targets[j] = genImg()

np.save("sun_image_train.npy", data)
np.save("sun_label_train.npy", targets)
        
# Save test data
samples = 1000
data = np.zeros([samples,28,28])
targets = np.zeros(samples, dtype=int)
for j in range(samples):
    data[j], targets[j] = genImg()

np.save("sun_image_test.npy", data)
np.save("sun_label_test.npy", targets)
```
